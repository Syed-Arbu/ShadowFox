```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")  # nicer default style
```

```python
np.random.seed(42)

months = np.arange(1, 13)
values = np.random.randint(20, 100, size=12)
category = np.random.choice(['A', 'B', 'C'], size=12)

df = pd.DataFrame({
    'Month': months,
    'Value': values,
    'Category': category
})

df.head()
```

```python
# Matplotlib
plt.figure(figsize=(8,4))
plt.plot(df['Month'], df['Value'], marker='o')
plt.title("Line Plot - Matplotlib")
plt.xlabel("Month")
plt.ylabel("Value")
plt.show()

# Seaborn
plt.figure(figsize=(8,4))
sns.lineplot(data=df, x="Month", y="Value", marker="o")
plt.title("Line Plot - Seaborn")
plt.show()
```

### Line Plot
- **Description**: Shows how a variable changes over time.
- **Use case**: Stock prices, sales per month.
- **Example**: We used months vs. values.


```python
# Matplotlib
plt.figure(figsize=(6,4))
plt.scatter(df['Month'], df['Value'], c='red')
plt.title("Scatter Plot - Matplotlib")
plt.xlabel("Month")
plt.ylabel("Value")
plt.show()

# Seaborn
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="Month", y="Value", hue="Category", s=100)
plt.title("Scatter Plot - Seaborn")
plt.show()
```

### Scatter Plot
- **Description**: Shows how a variable changes over time.
- **Use case**: Stock prices, sales per month.
- **Example**: We used months vs. values.


```python
# Matplotlib
counts = df['Category'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(counts.index, counts.values, color=['blue','green','orange'])
plt.title("Bar Chart - Matplotlib")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()

# Seaborn
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Category", palette="pastel")
plt.title("Bar Chart - Seaborn")
plt.show()
```

### Bar Chart
- **Description**: Shows how a variable changes over time.
- **Use case**: Stock prices, sales per month.
- **Example**: We used months vs. values.


```python
# Matplotlib only (Seaborn doesnt support pie)
sizes = counts.values
labels = counts.index
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart - Matplotlib")
plt.show()
```

### Pie Chart
- **Description**: Shows how a variable changes over time.
- **Use case**: Stock prices, sales per month.
- **Example**: We used months vs. values.


```python
# Matplotlib
plt.figure(figsize=(6,4))
plt.boxplot([df[df['Category']==c]['Value'] for c in df['Category'].unique()],
            labels=df['Category'].unique())
plt.title("Boxplot - Matplotlib")
plt.show()

# Seaborn
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="Category", y="Value", palette="Set2")
plt.title("Boxplot - Seaborn")
plt.show()
```

### Box Plot
- **Description**: Shows how a variable changes over time.
- **Use case**: Stock prices, sales per month.
- **Example**: We used months vs. values.


```python
plt.figure(figsize=(6,5))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap - Seaborn")
plt.show()
```

### Heatmap
- **Description**: Shows how a variable changes over time.
- **Use case**: Stock prices, sales per month.
- **Example**: We used months vs. values.


Matplotlib Overview

Oldest & most powerful plotting library in Python.

Full control over customization.

Used for scientific & academic publications.

Seaborn Overview

Built on Matplotlib but prettier by default.

Great for statistics (boxplots, histograms, heatmaps).

Less code for complex visualizations.

| Feature        | Matplotlib             | Seaborn                      |
| -------------- | ---------------------- | ---------------------------- |
| Ease of use    | More code, harder      | Easier, fewer lines          |
| Customization  | Very high              | Limited (but enough)         |
| Interactivity  | No (static)            | No (static)                  |
| Best use cases | Research, publications | Quick EDA, statistical plots |
